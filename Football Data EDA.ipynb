```python
# 4) Converting M to million K to thousand.

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

df['Wage'] = df['Wage'].apply(value_to_float)
```

```python
# Displaying the converted form of Wage column
df['Wage']
```

```python
# Displaying the converted form of Value column
df['Value'] = df['Value'].apply(value_to_float)
df['Value']
```

```python
# Displaying the converted form of Release Clause column
df['Release Clause'] = df['Release Clause'].apply(value_to_float)
df['Release Clause']
```

```python
# Now changing the datatype
df['Value'] = pd.to_numeric(df['Value'],downcast='float')
df.dtypes
```

```python
df['Release Clause'] = pd.to_numeric(df['Release Clause'],downcast='float')
df.dtypes
```

```python
# 9) Percentage missing values and imputation
percentage_missing_values = df.isnull().sum()*100 / len(df)
percentage_missing_values
```

1) Loaned From (has 93% missing values) is the club from where the player moves on a temporary basis hence it is a object datatype, it does not help us in any numeric analysis hence we can delete the column. 

2) Release clause has 8.59% values missing we can impute it with median value of the Release Clause. Rest all the columns have very less missing values and can be ignored

```python
# Dropping the "Loaned from" Column

df = df.drop(['Loaned From'], axis=1)
df
```

```python
# Imputing median value for "Release clause" column
df['Release Clause'] = df['Release Clause'].fillna(df['Release Clause'].median())
df.isnull().sum()/len(df)*100
```

Performed median imputation on Release Clause column
